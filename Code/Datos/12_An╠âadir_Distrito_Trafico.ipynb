# Calcular Trafico

```python
import pandas as pd

df_traf = pd.read_csv("datos/trafico2021_2023.csv")

df_traf
```

```python
df_traf = df_traf.drop(columns='Unnamed: 0')
df_traf
```

```python
df_ubic = pd.read_csv("datos/ubicaciones_sensores_trafico.csv")

df_ubic = df_ubic.drop(columns='Unnamed: 0')

df_ubic
```

# Asignar Distrito a las mediciones

```python
df_ubic = df_ubic.drop(columns=['tipo_elem', 'cod_cent', 'nombre', 'utm_x', 'utm_y', 'longitud', 'latitud'])

df_traf = pd.merge(df_traf, df_ubic, on='id', how='left')

df_traf
```

```python
df_traf['distrito'].unique()
```

# Eliminar datos incorrectos

```python
df_traf = df_traf[df_traf['error'] != 'E']
df_traf
#No ha eliminado nada
```

```python
df_traf.drop(columns='error', inplace=True)
df_traf
```

# Exportación de nuevo

```python
df_traf.to_csv('datos/trafico2021_2023.csv', index=False)
```
