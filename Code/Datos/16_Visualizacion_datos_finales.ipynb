# VISUALIZACIÓN DE LOS DATOS FINALES

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("datos/DATOS_FINALES.csv")

df.drop(columns='id', inplace=True)

df
```

```python
df['FECHA_HORA'] = pd.to_datetime(df['FECHA_HORA'])

df = df.sort_values(by='FECHA_HORA')

df_filtered = df[(df['ESTACION'] == 8) & (df['FECHA_HORA'].dt.strftime('%H:%M:%S') == '00:00:00')]

plt.figure(figsize=(10, 6)) 

plt.plot(df_filtered['FECHA_HORA'], df_filtered['TEMP'], linestyle='-')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
