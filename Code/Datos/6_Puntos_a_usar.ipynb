# Ver que puntos de medicion tienen mas datos.

```python
import pandas as pd

df = pd.read_csv("datos/cont_meteo_juntos_2021_2024.csv")

df.drop(columns='Unnamed: 0', inplace=True)
```

```python
import pandas as pd

columnas = ['NO', 'NO2', 'NOx', 'BEN', 'CO', 'EBE', 'O3', 'PM10', 'PM2_5', 'SO2', 'TOL']  # Puedes agregar más columnas aquí si es necesario

valores_nan_por_estacion = df.groupby('ESTACION')[columnas].apply(lambda x: x.isna().sum())

total_por_estacion = df['ESTACION'].value_counts().reset_index()
total_por_estacion.columns = ['ESTACION', 'TOTAL_MUESTRAS']

valores_nan_por_estacion = pd.merge(valores_nan_por_estacion, total_por_estacion, on='ESTACION')

for columna in columnas:
    valores_nan_por_estacion[columna] = (valores_nan_por_estacion[columna] / valores_nan_por_estacion['TOTAL_MUESTRAS']) * 100

valores_nan_por_estacion = valores_nan_por_estacion[['ESTACION'] + columnas]

valores_nan_por_estacion = valores_nan_por_estacion.sort_values(by='ESTACION').reset_index(drop=True)

valores_nan_por_estacion = valores_nan_por_estacion.round(2)

valores_nan_por_estacion
```

```python
df1 = pd.read_csv("datos/contaminacion/anio21/ene_mo21.csv", sep=';')
df2 = pd.read_csv("datos/contaminacion/anio21/feb_mo21.csv", sep=';')
df3 = pd.read_csv("datos/contaminacion/anio21/mar_mo21.csv", sep=';')
df4 = pd.read_csv("datos/contaminacion/anio21/abr_mo21.csv", sep=';')
df5 = pd.read_csv("datos/contaminacion/anio21/may_mo21.csv", sep=';')
df6 = pd.read_csv("datos/contaminacion/anio21/jun_mo21.csv", sep=';')
df7 = pd.read_csv("datos/contaminacion/anio21/jul_mo21.csv", sep=';')
df8 = pd.read_csv("datos/contaminacion/anio21/ago_mo21.csv", sep=';')
df9 = pd.read_csv("datos/contaminacion/anio21/sep_mo21.csv", sep=';')
df10 = pd.read_csv("datos/contaminacion/anio21/oct_mo21.csv", sep=';')
df11 = pd.read_csv("datos/contaminacion/anio21/nov_mo21.csv", sep=';')
df12 = pd.read_csv("datos/contaminacion/anio21/dic_mo21.csv", sep=';')

df2021 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

del [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

df1 = pd.read_csv("datos/contaminacion/anio22/ene_mo22.csv", sep=';')
df2 = pd.read_csv("datos/contaminacion/anio22/feb_mo22.csv", sep=';')
df3 = pd.read_csv("datos/contaminacion/anio22/mar_mo22.csv", sep=';')
df4 = pd.read_csv("datos/contaminacion/anio22/abr_mo22.csv", sep=';')
df5 = pd.read_csv("datos/contaminacion/anio22/may_mo22.csv", sep=';')
df6 = pd.read_csv("datos/contaminacion/anio22/jun_mo22.csv", sep=';')
df7 = pd.read_csv("datos/contaminacion/anio22/jul_mo22.csv", sep=';')
df8 = pd.read_csv("datos/contaminacion/anio22/ago_mo22.csv", sep=';')
df9 = pd.read_csv("datos/contaminacion/anio22/sep_mo22.csv", sep=';')
df10 = pd.read_csv("datos/contaminacion/anio22/oct_mo22.csv", sep=';')
df11 = pd.read_csv("datos/contaminacion/anio22/nov_mo22.csv", sep=';')
df12 = pd.read_csv("datos/contaminacion/anio22/dic_mo22.csv", sep=';')

df2022 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

del [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

df1 = pd.read_csv("datos/contaminacion/anio23/ene_mo23.csv", sep=';')
df2 = pd.read_csv("datos/contaminacion/anio23/feb_mo23.csv", sep=';')
df3 = pd.read_csv("datos/contaminacion/anio23/mar_mo23.csv", sep=';')
df4 = pd.read_csv("datos/contaminacion/anio23/abr_mo23.csv", sep=';')
df5 = pd.read_csv("datos/contaminacion/anio23/may_mo23.csv", sep=';')
df6 = pd.read_csv("datos/contaminacion/anio23/jun_mo23.csv", sep=';')
df7 = pd.read_csv("datos/contaminacion/anio23/jul_mo23.csv", sep=';')
df8 = pd.read_csv("datos/contaminacion/anio23/ago_mo23.csv", sep=';')
df9 = pd.read_csv("datos/contaminacion/anio23/sep_mo23.csv", sep=';')
df10 = pd.read_csv("datos/contaminacion/anio23/oct_mo23.csv", sep=';')
df11 = pd.read_csv("datos/contaminacion/anio23/nov_mo23.csv", sep=';')
df12 = pd.read_csv("datos/contaminacion/anio23/dic_mo23.csv", sep=';')

df2023 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

del [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

df = pd.concat([df2021, df2022, df2023])

del [df2021, df2022, df2023]

df['FECHA'] = df['DIA'].astype(str) + '/' + df['MES'].astype(str) + '/' + df['ANO'].astype(str)

df = df.drop(columns=['ANO', 'MES', 'DIA'])

df['FECHA'] = pd.to_datetime(df['FECHA'], format='%d/%m/%Y')

mapeo = {1: 'SO2',
         6: 'CO',
         7: 'NO',
         8: 'NO2',
         9: 'PM2_5',
         10: 'PM10',
         12: 'NOx',
         14: 'O3',
         20: 'TOL',
         30: 'BEN',
         35: 'EBE',
         37: 'MXY',
         38: 'PXY',
         39: 'OXY',
         42: 'TCH',
         43: 'CH4',
         44: 'NMHC'}

df['MAGNITUD'] = df['MAGNITUD'].map(mapeo)

df = df.drop(columns=['PROVINCIA', 'MUNICIPIO'])

import numpy as np

def eliminar_contenido(df, v_column, h_column):
    # Obtener los índices donde v es N
    indices_v_F = df.loc[df[v_column] == 'N'].index
    # Eliminar el valor de las celdas
    df.loc[indices_v_F, h_column] = np.nan
    return df

df = eliminar_contenido(df, 'V01', 'H01')
df = eliminar_contenido(df, 'V02', 'H02')
df = eliminar_contenido(df, 'V03', 'H03')
df = eliminar_contenido(df, 'V04', 'H04')
df = eliminar_contenido(df, 'V05', 'H05')
df = eliminar_contenido(df, 'V06', 'H06')
df = eliminar_contenido(df, 'V07', 'H07')
df = eliminar_contenido(df, 'V08', 'H08')
df = eliminar_contenido(df, 'V09', 'H09')
df = eliminar_contenido(df, 'V10', 'H10')
df = eliminar_contenido(df, 'V11', 'H11')
df = eliminar_contenido(df, 'V12', 'H12')
df = eliminar_contenido(df, 'V13', 'H13')
df = eliminar_contenido(df, 'V14', 'H14')
df = eliminar_contenido(df, 'V15', 'H15')
df = eliminar_contenido(df, 'V16', 'H16')
df = eliminar_contenido(df, 'V17', 'H17')
df = eliminar_contenido(df, 'V18', 'H18')
df = eliminar_contenido(df, 'V19', 'H19')
df = eliminar_contenido(df, 'V20', 'H20')
df = eliminar_contenido(df, 'V21', 'H21')
df = eliminar_contenido(df, 'V22', 'H22')
df = eliminar_contenido(df, 'V23', 'H23')
df = eliminar_contenido(df, 'V24', 'H24')

df = df.drop(columns=['V01', 'V02', 'V03', 'V04', 'V05', 'V06', 'V02', 'V08', 'V09', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21', 'V22', 'V23', 'V24'])

df = pd.melt(df, 
                id_vars=['ESTACION', 'MAGNITUD', 'PUNTO_MUESTREO', 'FECHA'], 
                value_vars=['H01', 'H02', 'H03', 'H04', 'H05', 'H06', 'H07', 'H08', 'H09', 'H10', 'H11', 'H12', 'H13', 'H14', 'H15', 'H16', 'H17', 'H18', 'H19', 'H20', 'H21', 'H22', 'H23', 'H24'], 
                var_name='HORA', 
                value_name='VALOR')

df
```
