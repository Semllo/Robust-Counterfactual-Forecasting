# Calcular trafico por distrito

```python
import pandas as pd

df = pd.read_csv("datos/trafico_calculado_intensidad_hora.csv")

df
```

## Hay que sumar el total de intensidad para cada hora y distrito

## Quedarse solo con urbano

```python
df['tipo_elem'].value_counts()
```

```python
df = df.loc[df['tipo_elem'] == 'URB']
df
```

# Calculo

```python
df_calculado = df.groupby(['distrito', 'fecha'])['intensidad'].sum().reset_index()

df_calculado
```

```python
df_calculado.to_csv('datos/trafico_calculado_por_distrito.csv', index=False)  # index=False para no incluir los índices como una columna adicional
```
