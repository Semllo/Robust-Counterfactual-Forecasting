## Eliminar datos no Validos

```python
import numpy as np

def eliminar_contenido(df, v_column, h_column):
    # Obtener los índices donde v es N
    indices_v_F = df.loc[df[v_column] == 'N'].index
    # Eliminar el valor de las celdas
    df.loc[indices_v_F, h_column] = np.nan
    return df
```

```python
df = eliminar_contenido(df, 'V01', 'H01')
df = eliminar_contenido(df, 'V02', 'H02')
df = eliminar_contenido(df, 'V03', 'H03')
df = eliminar_contenido(df, 'V04', 'H04')
df = eliminar_contenido(df, 'V05', 'H05')
df = eliminar_contenido(df, 'V06', 'H06')
df = eliminar_contenido(df, 'V07', 'H07')
df = eliminar_contenido(df, 'V08', 'H08')
df = eliminar_contenido(df, 'V09', 'H09')
df = eliminar_contenido(df, 'V10', 'H10')
df = eliminar_contenido(df, 'V11', 'H11')
df = eliminar_contenido(df, 'V12', 'H12')
df = eliminar_contenido(df, 'V13', 'H13')
df = eliminar_contenido(df, 'V14', 'H14')
df = eliminar_contenido(df, 'V15', 'H15')
df = eliminar_contenido(df, 'V16', 'H16')
df = eliminar_contenido(df, 'V17', 'H17')
df = eliminar_contenido(df, 'V18', 'H18')
df = eliminar_contenido(df, 'V19', 'H19')
df = eliminar_contenido(df, 'V20', 'H20')
df = eliminar_contenido(df, 'V21', 'H21')
df = eliminar_contenido(df, 'V22', 'H22')
df = eliminar_contenido(df, 'V23', 'H23')
df = eliminar_contenido(df, 'V24', 'H24')
```

# Normalizacion

### Pasar Hs a filas

```python
df = pd.melt(df, 
                id_vars=['ESTACION', 'MAGNITUD', 'PUNTO_MUESTREO', 'FECHA'], 
                value_vars=['H01', 'H02', 'H03', 'H04', 'H05', 'H06', 'H07', 'H08', 'H09', 'H10', 'H11', 'H12', 'H13', 'H14', 'H15', 'H16', 'H17', 'H18', 'H19', 'H20', 'H21', 'H22', 'H23', 'H24'], 
                var_name='HORA', 
                value_name='VALOR')

df
```

### Pasar MAGNITUD a columnas

```python
df['MAGNITUD'] = df['MAGNITUD'].astype(str)

df = df.pivot_table(index=['ESTACION', 'FECHA', 'HORA'], 
                            columns='MAGNITUD', 
                            values='VALOR',
                            aggfunc='first').reset_index()

df
```

```python
nan_percentage = (df.isna().mean() * 100).round(2)

print(nan_percentage)
```

## Pasar a formato numerico, si es necesario

```python
columnas_a_numerico = ['DIR_VIENT', 'HUM_REL', 'PRECIPITACION', 'PRES_BARIOMETRICA', 'RAD_SOL', 'TEMP', 'VEL_VIENT']

for columna in columnas_a_numerico:
    if df[columna].dtype == 'object':  
        df[columna] = df[columna].str.replace(',', '.')  
        df[columna] = pd.to_numeric(df[columna], errors='coerce')  

df
```

## Juntar Fecha y Hora

```python
df['FECHA'] = pd.to_datetime(df['FECHA'])

def convertir_hora(hora):
    if hora == 'H24':
        return '00:00'
    else:
        return hora[1:] + ':00'

df['HORA'] = df['HORA'].apply(convertir_hora)

df['FECHA_HORA'] = df['FECHA'].dt.strftime('%Y-%m-%d') + ' ' + df['HORA']

df['FECHA_HORA'] = pd.to_datetime(df['FECHA_HORA'])
#df_meteo = df_meteo.sort_values(by='FECHA_HORA')

df = df.drop(columns=['FECHA', 'HORA'])

df
```

## Reordenar y Exportar

```python
df = df[['ESTACION', 'FECHA_HORA', 'TEMP', 'HUM_REL', 'PRECIPITACION', 'PRES_BARIOMETRICA', 'RAD_SOL', 'DIR_VIENT', 'VEL_VIENT']]
df
```
