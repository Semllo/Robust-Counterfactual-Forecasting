# Juntar

## Primero restar una hora a los datos de trafico para poder juntarlos

Porque la medicion de las 17 de contaminación ha influido el trafico durante la hora 16


```python
# Convertimos la columna 'fecha' a tipo datetime
df_traf['fecha'] = pd.to_datetime(df_traf['fecha'])

# Restamos una hora a la columna 'fecha'
df_traf['fecha'] = df_traf['fecha'] - pd.Timedelta(hours=1)

df_traf
```

## Juntar


```python
df_traf['distrito'] = df_traf['distrito'].astype(int)
df_traf
```

```python
df_traf.dtypes
```

```python
df_meteo['FECHA_HORA'] = pd.to_datetime(df_meteo['FECHA_HORA'])
df_meteo.dtypes
```

```python
# Supongamos que df1 es tu DataFrame izquierdo y df2 es tu DataFrame derecho

df_traf = df_traf.rename(columns={'distrito': 'DISTRITO'})
df_traf = df_traf.rename(columns={'fecha': 'FECHA_HORA'})


df_final = pd.merge(df_meteo, df_traf, on=['DISTRITO', 'FECHA_HORA'], how='left')


df_final
```

```python
print(df_traf.loc[(df_traf['FECHA_HORA'] == '2021-01-01 03:00:00') & (df_traf['DISTRITO'] == 3)])
```

```python
df_final.to_csv('datos/DATOS_FINALES.csv', index=False)  # index=False para no incluir los índices como una columna adicional
```
