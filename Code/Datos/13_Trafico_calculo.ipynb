# Calcular Trafico

```python
import pandas as pd

df_traf = pd.read_csv("datos/trafico2021_2023.csv")

df_traf
```

## Primero eliminar duplicados.

```python
df_traf = df_traf.drop_duplicates().reset_index()
df_traf
```

Se reduce considerablemente el numero de filas

## Calcular la suma de intensidad (numero total de coches que pasa en 1h por ese punto)

```python
# Supongamos que 'df' es tu DataFrame
df_traf['fecha'] = pd.to_datetime(df_traf['fecha'])

# Eliminamos los minutos y segundos de la fecha
df_traf['fecha'] = df_traf['fecha'].apply(lambda x: x.replace(minute=0, second=0))

df_traf.set_index('fecha', inplace=True)


# CALCULO
# Agrupamos por id, tipo_elem, distrito y fecha, y sumamos la intensidad
df_calculado = df_traf.groupby(['id', 'tipo_elem', 'distrito', df_traf.index]).intensidad.sum().reset_index()

# Cambiamos el formato de la fecha para que muestre hasta los segundos
df_calculado['fecha'] = df_calculado['fecha'].dt.strftime('%Y-%m-%d %H:%M:%S')

df_calculado
```

```python
df_calculado.to_csv("datos/trafico_calculado_intensidad_hora.csv")
```

```python
df_reducido = df_traf.head(40)
df_reducido
```

```python
df_reducido = df_reducido.drop_duplicates().reset_index()
df_reducido
```

```python
df = df_reducido

# Supongamos que 'df' es tu DataFrame
df['fecha'] = pd.to_datetime(df['fecha'])

# Eliminamos los minutos y segundos de la fecha
df['fecha'] = df['fecha'].apply(lambda x: x.replace(minute=0, second=0))

df.set_index('fecha', inplace=True)

# Agrupamos por id, tipo_elem, distrito y fecha, y sumamos la intensidad
df_resumen = df.groupby(['id', 'tipo_elem', 'distrito', df.index]).intensidad.sum().reset_index()

# Cambiamos el formato de la fecha para que muestre hasta los segundos
df_resumen['fecha'] = df_resumen['fecha'].dt.strftime('%Y-%m-%d %H:%M:%S')

(df_resumen)
```
