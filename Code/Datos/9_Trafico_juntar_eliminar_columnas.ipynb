# DATOS TRAFICO 2021

```python
df1 = pd.read_csv("datos/trafico/mediciones/2021/01-2021.csv", sep=';')
df2 = pd.read_csv("datos/trafico/mediciones/2021/02-2021.csv", sep=';')
df3 = pd.read_csv("datos/trafico/mediciones/2021/03-2021.csv", sep=';')
df4 = pd.read_csv("datos/trafico/mediciones/2021/04-2021.csv", sep=';')
df5 = pd.read_csv("datos/trafico/mediciones/2021/05-2021.csv", sep=';')
df6 = pd.read_csv("datos/trafico/mediciones/2021/06-2021.csv", sep=';')
df7 = pd.read_csv("datos/trafico/mediciones/2021/07-2021.csv", sep=';')
df8 = pd.read_csv("datos/trafico/mediciones/2021/08-2021.csv", sep=';')
df9 = pd.read_csv("datos/trafico/mediciones/2021/09-2021.csv", sep=';')
df10 = pd.read_csv("datos/trafico/mediciones/2021/10-2021.csv", sep=';')
df11 = pd.read_csv("datos/trafico/mediciones/2021/11-2021.csv", sep=';')
df12 = pd.read_csv("datos/trafico/mediciones/2021/12-2021.csv", sep=';')

print(list(df1.columns) == list(df2.columns) == list(df3.columns) == list(df4.columns) == list(df5.columns) == list(df6.columns) == list(df7.columns) == list(df8.columns) == list(df9.columns) == list(df10.columns) == list(df11.columns) == list(df12.columns ))

dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

columnas_a_eliminar = ['ocupacion', 'carga', 'vmed', 'periodo_integracion']

for df in dfs:
    df.drop(columns=columnas_a_eliminar, inplace=True)

df_trafico_2021 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

df_trafico_2021 = df_trafico_2021.reset_index(drop=True)

del [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

df_trafico_2021
```
