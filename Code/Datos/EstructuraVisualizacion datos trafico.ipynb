# Organización datos de trafico

```python
import pandas as pd

df_historico = pd.read_csv("datos/trafico/pruebas/historico12-2023.csv", sep=';')
df_ubicaciones = pd.read_csv("datos/trafico/pruebas/ubicacion_12-2023.csv", sep=';')
```

## Coincidencias ids

```python
import numpy as np
```

```python
valores_unicos_hist = df_historico['id'].unique()
valores_unicos_ubic = df_ubicaciones['id'].unique()
```

```python
print(len(valores_unicos_hist))
print(len(valores_unicos_ubic))
```

Hay mas ubicaciones que en el historico

```python
interseccion = np.intersect1d(valores_unicos_hist, valores_unicos_ubic)

len(interseccion)
```

Bien, tengo todas las ubicaciones de los puntos donde se detecta el trafico.

## Mapa con las ubicaciones de los puntos de medida

```python
filtro = df_ubicaciones['id'].isin(interseccion)

df = df_ubicaciones[filtro]

df
```

```python
# Eliminar las 12 posiciones empezando por detras de la cadena, eliminar el punto

def recortar_cadena(cadena):
    return cadena[:-12]

df['longitud'] = df['longitud'].apply(recortar_cadena)
df['latitud'] = df['latitud'].apply(recortar_cadena)
```

```python
# Eliminar el punto

df['longitud'] = df['longitud'].apply(lambda x: x.replace('.', '') if isinstance(x, str) else x)

df['latitud'] = df['latitud'].apply(lambda x: x.replace('.', '') if isinstance(x, str) else x)
```

```python
# Colocar el punto donde toca

def colocar_punto(cadena):
    if len(cadena) >= 3:
        return cadena[:2] + '.' + cadena[2:]
    else:
        return cadena

# Aplicar la función lambda para colocar un punto en la posición 3 de la columna seleccionada
df['longitud'] = df['longitud'].apply(colocar_punto)
df['latitud'] = df['latitud'].apply(colocar_punto)

df
```

```python
import folium

# Crear un mapa centrado en unas coordenadas específicas
mapa = folium.Map(location=[40.41831, -3.70275], zoom_start=12)

for index, row in df.iterrows():
    folium.Marker(location=[row['latitud'], row['longitud']], popup=row['nombre']).add_to(mapa)


# Visualizar el mapa
mapa
```
