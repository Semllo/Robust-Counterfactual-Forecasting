# Juntar Datos Contaminacion - Meteo

```python
import pandas as pd

dfMETEO = pd.read_csv("datos/meteorologia2021_2023.csv", sep=',')
dfCONTAMINACION = pd.read_csv("datos/contaminacion2021_2023.csv", sep=',')
```

```python
dfMETEO = dfMETEO.drop(columns=['Unnamed: 0'])
dfCONTAMINACION = dfCONTAMINACION.drop(columns=['Unnamed: 0'])
```
